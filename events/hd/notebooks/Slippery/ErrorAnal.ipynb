```python
preds = pd.read_csv('/Users/rbekbolatov/Downloads/train_1804_10.csv')
preds.columns = ['pred']
data = pd.read_csv('/Users/rbekbolatov/data/kaggle/homedepot/train.csv')
data = preds.join(data).set_index('id')
data['diff'] = data['pred'] - data['relevance']
data_train_stack = data.loc[:,['diff', 'pred', 'relevance', 'search_term', 'product_title']]
data_train_stack.sort_values('diff', ascending=False, inplace=True)
```

```python
a = data_train_stack['diff'].sort_index().reset_index()
a.groupby(lambda r: math.ceil(r/100)).apply(lambda x: x.describe()).loc(axis=0)[:,'mean'].reset_index().plot(kind='scatter', x='level_0',  y='diff', s=0.2)
```
