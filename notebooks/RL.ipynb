```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

matplotlib.rcParams['figure.figsize'] = (8, 5)
```

```python
# update all predecessors

# depth 2 (1)
t = np.arange(100, 1001, 100)
a = np.array([1.61, 4.33, 7.18, 16.01, 16.57, 29.23, 38.83, 53.32, 54.57, 70.22])
s = np.array([-500.17, -500.64, -488.72, -123.75, -34.29, 86.50, 76.74, 136.84, 157.92, 296.82])

# depth 1 (0)
a1 = np.array([1.06, 2.30, 3.62, 4.97, 10.77, 16.85, 19.67, 22.49, 23.76, 28.80])
s1 = np.array([-499.84, -510.70, -419.83, -381.34, -115.78, 95.97, 26.69, 106.01, 35.60, 136.30])

# update only predecessor on current path:
a_base = np.array([0.61, 0.91, 0.98, 1.17, 1.23, 1.17, 1.21, 1.27, 1.28, 1.33])
s_base = np.array([-499.43, -512.30, -410.50, -372.09, -330.26, -279.31, -281.31, -220.92, -109.45, -18.72])
```

```python
plt.plot(t, a_base, 'b')
plt.plot(t, a1, 'g')
plt.plot(t, a, 'r')
plt.xlabel('#runs')
plt.ylabel('time per batch (seconds)')
plt.legend(['ony last (base)', 'depth 1','depth 2'], loc='upper left')
plt.title('Times per batch (100 runs)')
```

```python
plt.plot(t, s_base, 'b')
plt.plot(t, s1, 'g')
plt.plot(t, s, 'r')
plt.xlabel('#runs')
plt.ylabel('Avg rewards')
plt.legend(['only last (base)', 'depth 1','depth 2'], loc='upper left')
plt.title('Average Rewards (100 runs)')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(t, a, 'b')
ax1.set_ylabel('times')

ax2 = ax1.twinx()
ax2.plot(t, s, 'r-')
ax2.set_ylabel('scores', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')
```
