```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
from datetime import datetime

plt.rcParams['figure.figsize'] = (13.0, 8.0)
```

```python
logs_passed = pd.read_csv('/Users/rbekbolatov/Downloads/tmp/logs_passed_2.csv')
logs_blocked = pd.read_csv('/Users/rbekbolatov/Downloads/tmp/logs_blocked_2.csv')
weblogs = logs_passed.append(logs_blocked)
weblogs['event_date'] = pd.to_datetime(weblogs['event_date'])
```

```python
weblogs_over_time = weblogs.pivot_table('cnt', columns=['state', 'event_type'], index='event_date', aggfunc='sum')
weblogs_over_time = weblogs_over_time.fillna(0)
```

```python
def event_over_time(event_type):
    if event_type in weblogs_over_time['passed'].columns and event_type in weblogs_over_time['blocked'].columns:
        return pd.DataFrame({
                'passed': weblogs_over_time['passed', event_type], 
                'blocked': weblogs_over_time['blocked', event_type], 
                'total': weblogs_over_time['passed', event_type] + weblogs_over_time['blocked', event_type]
                })
    if event_type in weblogs_over_time['passed'].columns:
        return pd.DataFrame({
                'passed': weblogs_over_time['passed', event_type], 
                'blocked': None, 
                'total': weblogs_over_time['passed', event_type]
                })
    if event_type in weblogs_over_time['blocked'].columns:
        return pd.DataFrame({
                    'passed': None, 
                    'blocked': weblogs_over_time['blocked', event_type],
                    'total': weblogs_over_time['blocked', event_type]
                    })
    raise KeyError('Event type "%s" was not found.' % str(event_type))
```

```python
all_event_types = list(weblogs_over_time.columns.levels[1])
my_event_types = ['page_view', 'ad_request', 'ad_impression', 'ad_click', 'contact_impression', 'professional_impression']
```

```python
all_events_over_time = { 
    event_type: event_over_time(event_type)
    for event_type in all_event_types
    }
```

```python
def plot_event_history(event_types, kind='total', start_month=5, start_day=1):
    num_event_types = len(event_types)
    num_cols = min(3, int(math.floor(math.sqrt(num_event_types))))
    num_rows = int(math.ceil(num_event_types*1.0/num_cols))
    if num_cols >= 3:
        width = 5
    elif num_cols == 2:
        width = 8
    else:
        width = 15
    plt.suptitle(kind)
    plt.figure(figsize=(num_cols*width, num_rows*5))
    for i, event_type in enumerate(event_types, 1):
        history = event_over_time(event_type)
        plt.subplot(num_rows, num_cols, i)
        #history[kind].ix[datetime(2015, start_month, start_day):].plot(style='b-', label=event_type)
        history.ix[datetime(2015, start_month, start_day):, kind].plot(style='b-', 
                                                                                      label=event_type,
                                                                                      stacked=True
                                                                                     )
        plt.legend()
```

```python
plot_event_history(my_event_types, 'total', 8)
```
