```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# http://docs.scipy.org/doc/numpy-1.10.1/reference/generated/numpy.random.exponential.html
N = 100
t = [np.random.exponential(scale=1.0, size=20) for i in range(N)]
for i in range(N):
    plt.plot(t[i].cumsum(),  color='gray',lw=0.04)
plt.show()
```
