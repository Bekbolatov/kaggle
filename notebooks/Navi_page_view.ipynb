# page_view by page_type

```sql
select event_date, page_type, count(1) from src.page_view where event_date >= '2015-09-30' group by event_date, page_type;
```

````python
acv_navi = pd.read_csv(base_loc + 'page_view_page_type_navi.csv')
acv_navi['system'] = 'Navi'
acv_navi.rename(columns={'_c2': 'cnt'}, inplace=True)

acv_prod = pd.read_csv(base_loc + 'page_view_page_type_prod.csv')
acv_prod['system'] = 'Prod'
acv_prod.rename(columns={'_c2': 'cnt'}, inplace=True)

acv = pd.concat([acv_navi, acv_prod]).pivot_table('cnt', columns=['system'], index=['event_date', 'page_type']).sort_index()
acv['diff_abs'] = acv['Navi'] - acv['Prod']
acv['diff'] = acv['diff_abs'] / acv['Prod']
acv = acv[['Prod', 'Navi', 'diff_abs', 'diff']]
acv
````
