```python
base_loc = '/Users/rbekbolatov/'
loc = base_loc + 'data/wesmckinney/names/'
years = range(1880, 2011)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = loc + 'yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    frame['year'] = year
    pieces.append(frame)

names = pd.concat(pieces, ignore_index=True)
```

```python
ssl_ad_req = pd.read_csv('/Users/rbekbolatov/Documents/ssl_counts_ad_request_apr07.csv')
ssl_ad_req_data = ssl_ad_req.set_index(['template_name', 'event_date']).sort_index()
datas = ssl_ad_req_data

from pylab import rcParams
rcParams['figure.figsize'] = 25, 20


els = datas.reset_index()['template_name'].unique()
els = [x for x in els if x != 'Lawyer Profile - Resume']

xxx = [(datas.loc[x]['cnt'].sum() , x) for x in els]
els = [y[1] for y in sorted(xxx, reverse=True) if y[0] > 2200]

aabb = len(els)
import math
aa = 3 #math.ceil(math.sqrt(aabb))
bb = math.ceil(aabb*1.0/aa)

fig = plt.figure()
st = fig.suptitle("Unfiltered event counts (hourly)", fontsize="x-large")
for i, event_type in enumerate(els):
    if event_type == 'Lawyer Profile - Resume':
        continue
    ax = fig.add_subplot(bb, aa, i + 1)
    for event_date in datas.loc[event_type].index.unique():
        dat = datas.loc[(event_type, event_date)].sort_values(by='hh')
        ax.plot(dat['hh'], dat['cnt'], label = "{0}".format(event_date))
    plt.xlabel("Hour")
    plt.ylabel("Counts")
    plt.title(event_type)
    ax.legend()
plt.show()

st.set_y(0.95)
fig.subplots_adjust(top=0.85)
#plt.tight_layout(pad=0.2, w_pad=0.5, h_pad=5.0)
#plt.tight_layout()
```
